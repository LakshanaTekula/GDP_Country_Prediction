```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def regression_metrics(ytest, ytest_pred):

    RMSE = mean_squared_error(ytest,ytest_pred,squared=False)
    MSE = mean_squared_error(ytest,ytest_pred,squared=True)
    MAE = mean_absolute_error(ytest, ytest_pred)
    R2_score = r2_score(ytest,ytest_pred)
    Adj_R2_score = 1-(1-R2_score)*(60-1)/(60-len(df.columns)-2)

    print("RMSE:",RMSE)
    print("MSE:", MSE)
    print("MAE", MAE)
    print("R2_score:", R2_score)
    print("Adj_R2_score:", Adj_R2_score)
    return
```

```python
def loss(y,y_predicted):
    n=len(y)
    s=0
    for i in range(n):
        s+=(y[i]-y_predicted[i])**2
    return (1/n)*s
```

```python
def predicted_y(weight,x,intercept):
    y_lst=[]
    for i in range(len(x)):
        y_lst.append(weight@x[i]+intercept) #dot product btween weight and input
    return np.array(y_lst)
```

```python
def dldw(x,y,y_predicted):
    s=0
    n=len(y)
    for i in range(n):
        s+=-x[i]*(y[i]-y_predicted[i])
    return (2/n)*s
```

```python
def dldb(y,y_predicted):
    n=len(y)
    s=0
    for i in range(len(y)):
        s+=-(y[i]-y_predicted[i])
    return (2/n) * s
```

```python
def gradient_descent(x,y):

    weight_vector=np.random.randn(x.shape[1])
    intercept=0
    epoch = 3000

    n = len(x)
    linear_loss=[]
    learning_rate = 0.001

    for i in range(epoch):


        y_predicted = predicted_y(weight_vector,x,intercept)

        weight_vector = weight_vector - learning_rate *dldw(x,y,y_predicted)


        intercept = intercept - learning_rate * dldb(y,y_predicted)
        linear_loss.append(loss(y,y_predicted))

    plt.plot(np.arange(1,epoch),linear_loss[1:])
    plt.xlabel("number of epoch")
    plt.ylabel("loss")

    return weight_vector,intercept
```

```python
w,b=gradient_descent(X_train_transform,y_train)

print("weight:",w)
print("bias:",b)
```

```python
def predict(inp):
    y_lst=[]
    for i in range(len(inp)):
        y_lst.append(w@inp[i]+b)
    return np.array(y_lst)
```

```python
y_pred_manual=predict(X_train_transform)
y_pred_manual
```

```python
df3_pred=pd.DataFrame()
df3_pred["y_actual"]=y_train
df3_pred["y_predicted"]=np.round(y_pred_manual,1)
df3_pred
```

```python
regression_metrics(y_test, predict(X_test_transform))
```
